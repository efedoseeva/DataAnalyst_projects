**Поиск аномальных значений для непрерывных признаков**

```python
#проверим аномальные значения
def calc_boxplot(df_col : pd.Series) -> tuple:
    """
    Функция для расчета значений нижнего и верхнего усов.
    На вход подается столбец датафрейма,
    на выходе кортеж: первый элемент -- нижний ус, второй элемент -- верхний ус
    """
    Q1, median, Q3 = np.percentile(np.asarray(df_col.dropna()), [25, 50, 75])
    IQR = Q3 - Q1
    loval = Q1 - 1.5 * IQR
    hival = Q3 + 1.5 * IQR
    wiskhi = np.compress(np.asarray(df_col.dropna()) <= hival, np.asarray(df_col.dropna()))
    wisklo = np.compress(np.asarray(df_col.dropna()) >= loval, np.asarray(df_col.dropna()))
    actual_hival = np.max(wiskhi)
    actual_loval = np.min(wisklo)
    return actual_loval, actual_hival #(нижний ус, верхний ус)
```

```python
# построим диаграмму размаха
plt.subplots(3, figsize=(24, 36))
for index_fig, col in enumerate(lst_cont, start=0):
    plt.subplot(3, 2, 2*index_fig + 1)
    ax = sns.boxplot(data=df,
                 y=col)
    actual_loval, actual_hival = calc_boxplot(df[col])
    ax.axhline(actual_loval,
               color='blue',
               label=f'Нижный ус -- {actual_loval}')
    ax.axhline(actual_hival, color='red',
               label=f'Верхний ус -- {actual_hival}')
    q_99 = np.percentile(df[col].dropna(), 99)
    ax.axhline(q_99,
               color='green',
               label=f'99-я перцинтель -- {q_99:.2f}')
    q_1 = np.percentile(df[col].dropna(), 1)
    ax.axhline(q_1, color='yellow',
               label=f'1-я перцинтель -- {q_1:.2f}')
    plt.legend()
    plt.subplot(3, 2, 2*index_fig + 2)
    ax = sns.histplot(df[col]);
    q_99 = np.percentile(df[col].dropna(), 99)
    ax.axvline(q_99,
            color='green',
            label=f'99-я перцинтель -- {q_99:.2f}')
    q_1 = np.percentile(df[col].dropna(), 1)
    ax.axvline(q_1,
            color='yellow',
            label=f'1-я перцинтель -- {q_1:.2f}')
    actual_loval, actual_hival = calc_boxplot(df[col])
    ax.axvline(actual_loval,
               color='blue',
               label=f'Нижный ус -- {actual_loval}')
    ax.axvline(actual_hival, color='red',
               label=f'Верхний ус -- {actual_hival}')
    plt.legend()
```

```python
#создадим новый датафрейм только с топовыми 5 вакансиями и нужными признаками
list_top5=['повар', 'подсобный рабочий', 'уборщик служебных помещений',
            'бухгалтер', 'водитель автомобиля']

df_top5vacancies=df_new[['vacancy_name', 'education', 'education_speciality', 'required_experience', 'position_responsibilities', 'hard_skills',
       'soft_skills']]
df_top5vacancies = df_top5vacancies.loc[df_top5vacancies['vacancy_name'].isin(list_top5)]
df_top5vacancies.head(2)
```

```python
#добавим 2 новых столбца в которых hard_skills	soft_skills распарсены в список без лишней информации

# str=df_top10vacancies.hard_skills.loc[219]

def get_skill(str, substring):
    str = str.replace('"', '')
    skills = []
    index = -1  # начинаем поиск с начала строки

    while True:
        # находим следующее вхождение подстроки
        index = str.find(substring, index + 1)
        if index != -1:
            str1=str[index+16:]
            str2 = str1[:str1.find(',')]
        # если вхождение не найдено, выходим из цикла
        else:
            break

        # добавляем индекс в список
        skills.append(str2)

    return skills
```

```python
df_top5vacancies['hard_skills_exact']=df_top5vacancies['hard_skills'].apply(get_skill, args=('hard_skill_name',))
df_top5vacancies['soft_skills_exact']=df_top5vacancies['soft_skills'].apply(get_skill, args=('soft_skill_name',))
#избавимся от старых столбцов
# position_responsibilities текстовый признак, его сложно стандартизировать, поэтому исключаю его из анализа
df_top5vacancies=df_top5vacancies[['vacancy_name', 'education', 'education_speciality', 'required_experience',
                                   'hard_skills_exact', 'soft_skills_exact']]
df_top5vacancies.reset_index()
df_top5vacancies.loc[219]
```

Отдельно проанализируем софт и хард скилы

```python
#преобразование данных
#оставим только строки в которых есть скилы
df_top5vacancies_сut=df_top5vacancies[(df_top5vacancies['hard_skills_exact'].str.len()>2) | (df_top5vacancies['soft_skills_exact'].str.len()>2)]
```

```python
#Теперь выберем по топ 5 самых популярных soft и hard скилов
def get_skills_dct(skills_list):
    string = ''
    for el in skills_list:
        s=str(el).replace(' ', '')
        s = s.replace('[', '')
        s = s.replace(']', '')
        s = s.replace("'", '')
        string += s+','

    lst=string.split(',')

    lst = [i for i in lst if i]
    dct_ss=dict()

    for i in lst:
        dct_ss.update([(i,lst.count(i))])

    return sorted(dct_ss.items(), key=lambda item: item[1], reverse=True)[:5]

soft_skills_list=list(df_top5vacancies_сut.soft_skills_exact)
hard_skills_list=list(df_top5vacancies_сut.hard_skills_exact)
print(f'''Топ 5 soft skills по топовым вакансиям:
{get_skills_dct(soft_skills_list)}

Топ 5 hard skills по топовым вакансиям:
{get_skills_dct(hard_skills_list)}
      ''')
```

```python
#визуализируем результат

pd_ss=pd.DataFrame.from_dict(get_skills_dct(soft_skills_list))
pd_ss.columns=['skill','cnt']
pd_hs=pd.DataFrame.from_dict(get_skills_dct(hard_skills_list))
pd_hs.columns=['skill','cnt']


plt.subplots(2, figsize=(18, 8))
plt.subplots_adjust(wspace=0.1,
                    hspace=0.5)

plt.subplot(1, 2, 1)
ax=plt.pie(pd_ss.cnt, labels=pd_ss.skill)

plt.title('SOFT SKILLS')

plt.subplot(1, 2, 2)
ax=plt.pie(pd_hs.cnt, labels=pd_hs.skill)

plt.title('HARD SKILLS')

plt.show()
```

**Провека гипотезы** проверить гипотезу о равенстве средней заработной платы по отраслям.

Будем проверять попарно области, начнем с проверки какие сферы имеют нормальное распределение.


Проверим у какой сферы по минимальным зарплатам нормальное распределение
Применяем критерий Колмагорова-Смирнова, чтобы определить нормальный ли закон распределения у величин

```python

from scipy import stats
alpha=0.05
prof_list=list(df_new.professional_sphere_name.unique())
for i, col1 in enumerate(prof_list, start=0):
    # res = stats.shapiro(df_new[df_new.professional_sphere_name == col1].salary_min)
    res2=kstest(df_new[df_new.professional_sphere_name == col1].salary_min, 'norm')
    if (res2.pvalue>=alpha):
        print(f'''p-value={res2.pvalue} - {col1}''')
```

Код ничего не вывел, значит все величины имеют НЕ нормальное распределение

Применяем критерий Манна-Уитни и Краскела-Уоллиса для проверки нулевых гипотез о том, что средние зарплаты для каждой пары признаков не отличаются

```python

ind=0
alpha=0.05
s=set()
prof_list=list(df_new.professional_sphere_name.unique())
for i, col1 in enumerate(prof_list, start=0):
    prof_list.remove(col1)
    for j, col2 in enumerate(prof_list, start=0):
        ind+=1
        res1 = stats.mannwhitneyu(df_new[df_new.professional_sphere_name == col1].salary_min,
             df_new[df_new.professional_sphere_name == col2].salary_min)
        res2 = stats.kruskal(df_new[df_new.professional_sphere_name == col1].salary_min,
             df_new[df_new.professional_sphere_name == col2].salary_min)
        if (res1.pvalue>alpha):
            s.add(col1)
            s.add(col2)
            print(f'''
            Н_0: средняя зарплата из сферы '{col1}' и '{col2}' не отличаются
            Н_1: средняя зарплата из сферы '{col1}' и '{col2}' отличаются
            p-value={res1.pvalue} > {alpha}  значит гипотезу Н_0 принимаем!
            По коэффициенту Краскела-Уоллиса p-value={res2.pvalue}
                  ''')
print (f'''Итого {len(s)} сфер имеют одинаковую среднюю зарплату,
среди них: {s} ''')
```
